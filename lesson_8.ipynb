# Matplotlib & Seaborn

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x * 5
z = x ** 2
```

```python
fig = plt.figure()
ax = fig.add_axes( (0,0,1,1) )

ax.plot( x, y )
ax.set_xlabel('X asis')
ax.set_ylabel('Y asis')
ax.set_title('Pavadinimas')
```

```python
fig = plt.figure()
ax = fig.add_axes( (0,0,1,1) )
print(z[50])

z[49] = 2500
z[50] = 2200

ax.plot( x, z )

ax.plot(x, z)
ax_2 = fig.add_axes( (0.3,0.5,0.2,0.2) )
ax_2.plot(x,z)
ax_2.set_xlim(46, 54)
ax_2.set_ylim(2100, 2600)
ax_2.spines[['right', 'top']].set_visible(False)
```

```python
fig, axes = plt.subplots(3, 3)

for row in axes:
    row[2].plot(z,y)
    for col in row:
        col.plot(x,y)
```
